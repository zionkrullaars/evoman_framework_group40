```python
import math

np.random.seed(42)

def transform_rank(fitnesses, c):
    # Sort fitnesses in reverse order (higher fitness first)
    population_size = fitnesses.shape[0]
    ranks = np.argsort(-np.array(fitnesses))
    rank_scaling = []
    
    for i in range(population_size):
        x_i = np.random.rand()  # Random number in [0,1]
        # exp_term = np.exp(c * x_i) / (population_size * np.exp(c))
        exp_term = (population_size - population_size*math.e**(-c*x_i)) / (1-math.e**(-c))
        n_i = math.ceil(exp_term)
        rank_scaling.append((n_i, ranks[i]))
    
    # rank_scaling = np.array(rank_scaling)
    # assert fitnesses.shape == rank_scaling.shape, f"Fitnesses shape {fitnesses.shape} does not match rank scaling shape {rank_scaling.shape}"
    rank_scaling = sorted(rank_scaling, key=lambda x: x[0])
    return rank_scaling


fitness = np.sort(np.random.random_sample(100))
c = 0.02
ranks = transform_rank(fitness, c)
for rank in ranks:
    print(rank)
```
